```python
# ============================================================
# Morphology-Aware Bengali NER
# Latent Variable CRF: Morphological Segmentation as Latent Variable
#
# ── Formal Graphical Model ───────────────────────────────────
#
#   P(y, z | x) = (1/Z(x)) * prod_t phi(y_t, y_{t-1}, x_t, z_t) * P(z_t | x_t)
#
#   x_t = observed token
#   z_t = LATENT suffix-length in {0,1,2,3,4}
#   y_t = observed NER label
#
#   Marginal: P(y | x) = sum_z P(y, z | x)
#
# ── Models ───────────────────────────────────────────────────
#   1. Basic CRF        — word-level features only
#   2. Latent Morph CRF — learned morphology via EM
#   3. BiLSTM           — neural baseline
#
# ── Ablation Study ───────────────────────────────────────────
#   Measures contribution of each feature group to Latent CRF
#   by removing one group at a time (at 10% training data).
#
# ── References ───────────────────────────────────────────────
#   Lafferty et al. (2001) CRF
#   Quattoni et al. (2007) Hidden-state CRFs
#   Peng & McCallum (2004) Latent variable IE
#   Dempster et al. (1977) EM
# ============================================================

# pip install datasets sklearn-crfsuite seqeval torch tqdm matplotlib

import random
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from collections import defaultdict
from tqdm import tqdm

from datasets import load_dataset
from sklearn_crfsuite import CRF
from seqeval.metrics import precision_score, recall_score, f1_score, accuracy_score

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import (AutoTokenizer, AutoModelForTokenClassification,
                          get_linear_schedule_with_warmup)

# ── Seed & Device ─────────────────────────────────────────────

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using Device:", DEVICE)

# ── Label Space ───────────────────────────────────────────────

LABELS = {
    0: 'O',
    1: 'B-PER', 2: 'I-PER',
    3: 'B-ORG', 4: 'I-ORG',
    5: 'B-LOC', 6: 'I-LOC'
}
LABEL2ID  = {v: k for k, v in LABELS.items()}

# ── Data ──────────────────────────────────────────────────────

dataset = load_dataset("wikiann", "bn")

def convert_to_sentences(split):
    return [
        list(zip(item["tokens"], [LABELS[t] for t in item["ner_tags"]]))
        for item in split
    ]

train_sentences = convert_to_sentences(dataset["train"])
test_sentences  = convert_to_sentences(dataset["test"])
print(f"Train: {len(train_sentences)} | Test: {len(test_sentences)}")


# ============================================================
# MORPHOLOGICAL SEGMENTER
# ============================================================
#
# Models P(z_t | x_t) — the prior over suffix length for each token.
#
# Key design decisions:
#   1. Default log-weight = -2.0 (not 0.0).
#      With 0.0 default, every unseen suffix scores exp(0)=1.0,
#      dominating any trained weight — the segmenter never learns.
#      With -2.0, unseen suffixes score exp(-2)~0.135, so M-step
#      updates can meaningfully raise informative suffixes above background.
#
#   2. Within-bucket normalization in M-step.
#      Prevents empty suffix (length 0, present in every word) from
#      dominating via frequency rather than NER signal.
#
#   3. Laplace smoothing (alpha=0.1) on soft counts.
#      Stabilizes EM: prevents zero-count suffixes collapsing to -inf
#      log-weight in early iterations when CRF confidence is low.
# ============================================================

MAX_SUFFIX_LEN = 4
SUFFIX_LENGTHS  = list(range(MAX_SUFFIX_LEN + 1))
LAPLACE_ALPHA   = 0.1


def get_possible_segmentations(word):
    """All (stem, suffix) pairs; stem is never empty."""
    segs = []
    for slen in SUFFIX_LENGTHS:
        if slen == 0:
            segs.append((word, ""))
        elif slen < len(word):
            segs.append((word[:-slen], word[-slen:]))
    return segs


class MorphSegmenter:

    def __init__(self, default_log_weight=-2.0):
        self.default_log_weight  = default_log_weight
        self.log_suffix_weights  = defaultdict(lambda: default_log_weight)
        self._seen_suffixes      = set()

    def posterior(self, word):
        """P(z_t | x_t): softmax over candidate suffix log-weights.
        Returns {suffix_len: probability}."""
        segs   = get_possible_segmentations(word)
        scores = {len(sfx): np.exp(self.log_suffix_weights[sfx])
                  for _, sfx in segs}
        total  = sum(scores.values()) + 1e-12
        return {k: v / total for k, v in scores.items()}

    def most_likely_segmentation(self, word):
        """MAP: argmax_z P(z | x_t)."""
        post      = self.posterior(word)
        best_slen = max(post, key=post.get)
        for stem, suffix in get_possible_segmentations(word):
            if len(suffix) == best_slen:
                return stem, suffix
        return word, ""

    def m_step_update(self, soft_suffix_counts):
        """
        Update log_suffix_weights from expected sufficient statistics.
        Applies Laplace smoothing then normalizes within each length bucket.
        """
        # Laplace smoothing
        smoothed = {sfx: cnt + LAPLACE_ALPHA
                    for sfx, cnt in soft_suffix_counts.items()}

        # Group by suffix length
        by_length = defaultdict(dict)
        for suffix, count in smoothed.items():
            by_length[len(suffix)][suffix] = count

        # Normalize within each bucket
        for slen, bucket in by_length.items():
            total = sum(bucket.values()) + 1e-12
            for suffix, count in bucket.items():
                self.log_suffix_weights[suffix] = np.log(count / total + 1e-12)
                self._seen_suffixes.add(suffix)


# ============================================================
# LATENT MORPH CRF FEATURES
# ============================================================
#
# Feature groups (each can be ablated independently):
#   MORPH:   map.stem, map.suffix, top1.suffix, top2.suffix,
#            top1.prob, morph.entropy
#   NGRAMS:  prefix-2, prefix-3, suffix-2, suffix-3
#   CONTEXT: -1:word, -1:map.suffix, -1:suffix-3,
#            +1:word, +1:map.suffix, +1:suffix-3
#   WORD:    word, word.lower, word.length
#   MISC:    bias, is_digit, contains_hyphen, contains_number
# ============================================================

def word2features_latent(sent, i, segmenter, ablate=None):
    """
    Build CRF feature dict for token i.

    ablate: set of feature group names to remove, e.g. {"morph", "ngrams"}.
            None = full model (no ablation).
    """
    if ablate is None:
        ablate = set()

    word              = sent[i][0]
    stem, map_suffix  = segmenter.most_likely_segmentation(word)
    post              = segmenter.posterior(word)

    probs   = np.array(list(post.values()))
    entropy = -np.sum(probs * np.log(probs + 1e-12))

    suffix_probs = sorted(
        [(sfx, post.get(len(sfx), 0.0))
         for _, sfx in get_possible_segmentations(word)],
        key=lambda x: -x[1]
    )
    top1_suffix = suffix_probs[0][0] if suffix_probs else ""
    top2_suffix = suffix_probs[1][0] if len(suffix_probs) > 1 else ""
    top1_prob   = suffix_probs[0][1] if suffix_probs else 0.0

    features = {"bias": 1.0}

    # ── WORD features ─────────────────────────────────────────
    if "word" not in ablate:
        features.update({
            "word":       word,
            "word.lower": word.lower(),
            "word.length": len(word),
        })

    # ── MORPH features ────────────────────────────────────────
    if "morph" not in ablate:
        features.update({
            "map.stem":      stem,
            "map.suffix":    map_suffix,
            "top1.suffix":   top1_suffix,
            "top2.suffix":   top2_suffix,
            "top1.prob":     round(top1_prob, 2),
            "morph.entropy": round(entropy, 2),
        })

    # ── NGRAM features ────────────────────────────────────────
    if "ngrams" not in ablate:
        features.update({
            "prefix-2": word[:2],
            "prefix-3": word[:3],
            "suffix-2": word[-2:],
            "suffix-3": word[-3:],
        })

    # ── MISC features ─────────────────────────────────────────
    features.update({
        "is_digit":        word.isdigit(),
        "contains_hyphen": "-" in word,
        "contains_number": bool(re.search(r'\d', word)),
    })

    # ── CONTEXT features ──────────────────────────────────────
    if "context" not in ablate:
        if i > 0:
            pw   = sent[i-1][0]
            _, psuf = segmenter.most_likely_segmentation(pw)
            features.update({
                "-1:word":       pw,
                "-1:map.suffix": psuf,
                "-1:suffix-3":   pw[-3:],
            })
        else:
            features["BOS"] = True

        if i < len(sent) - 1:
            nw   = sent[i+1][0]
            _, nsuf = segmenter.most_likely_segmentation(nw)
            features.update({
                "+1:word":       nw,
                "+1:map.suffix": nsuf,
                "+1:suffix-3":   nw[-3:],
            })
        else:
            features["EOS"] = True
    else:
        # Always add BOS/EOS even when context ablated
        if i == 0:
            features["BOS"] = True
        if i == len(sent) - 1:
            features["EOS"] = True

    return features


# ============================================================
# EM TRAINING
# ============================================================

def em_train_latent_crf(train_sentences, n_em_iters=3, ablate=None):
    """
    Approximate EM for the Latent Morphology CRF.

    E-step: ner_confidence_t = P(true_label_t | x ; theta_crf)
            obtained from predict_marginals() — the posterior probability
            of the correct NER label under the current CRF.

    M-step (CRF):       retrain on features from current segmenter.
    M-step (Segmenter): normalize soft counts within suffix-length buckets.

    Returns: (crf, segmenter)
    """
    if ablate is None:
        ablate = set()

    segmenter = MorphSegmenter(default_log_weight=-2.0)
    crf       = None

    for em_iter in range(n_em_iters):
        print(f"    EM iter {em_iter+1}/{n_em_iters}", end=" ")

        # E-step: build features under current segmenter
        X_train, y_train = [], []
        for sent in train_sentences:
            X_train.append([word2features_latent(sent, i, segmenter, ablate)
                            for i in range(len(sent))])
            y_train.append([lbl for _, lbl in sent])

        # M-step (CRF)
        crf = CRF(algorithm='lbfgs', c1=0.1, c2=0.1,
                  max_iterations=100, all_possible_transitions=True)
        crf.fit(X_train, y_train)

        # E-step: CRF label marginals
        # predict_marginals(X)[sent_idx][token_idx] = {label: prob}
        y_marginals        = crf.predict_marginals(X_train)
        soft_suffix_counts = defaultdict(float)

        for sent, tok_marg_list in zip(train_sentences, y_marginals):
            for i, (token, true_label) in enumerate(sent):
                ner_conf   = tok_marg_list[i].get(true_label, 1e-12)
                morph_post = segmenter.posterior(token)
                for stem, suffix in get_possible_segmentations(token):
                    p_morph = morph_post.get(len(suffix), 0.0)
                    soft_suffix_counts[suffix] += p_morph * ner_conf

        # M-step (Segmenter)
        segmenter.m_step_update(soft_suffix_counts)
        print("done")

    return crf, segmenter


# ============================================================
# BASIC CRF BASELINE
# ============================================================

def word2features_basic(sent, i):
    word = sent[i][0]
    features = {
        "bias":            1.0,
        "word":            word,
        "word.lower":      word.lower(),
        "word.length":     len(word),
        "prefix-2":        word[:2],
        "prefix-3":        word[:3],
        "suffix-2":        word[-2:],
        "suffix-3":        word[-3:],
        "is_digit":        word.isdigit(),
        "contains_hyphen": "-" in word,
        "contains_number": bool(re.search(r'\d', word)),
    }
    if i > 0:
        features["-1:word"] = sent[i-1][0]
    else:
        features["BOS"] = True
    if i < len(sent) - 1:
        features["+1:word"] = sent[i+1][0]
    else:
        features["EOS"] = True
    return features


def prepare_data_basic(sentences):
    X = [[word2features_basic(s, i) for i in range(len(s))] for s in sentences]
    y = [[lbl for _, lbl in s] for s in sentences]
    return X, y


def train_crf(X_train, y_train):
    crf = CRF(algorithm='lbfgs', c1=0.1, c2=0.1,
              max_iterations=100, all_possible_transitions=True)
    crf.fit(X_train, y_train)
    return crf


# ============================================================
# BILSTM BASELINE
# ============================================================

word_vocab = {"<PAD>": 0, "<UNK>": 1}
for sent in train_sentences:
    for token, _ in sent:
        if token not in word_vocab:
            word_vocab[token] = len(word_vocab)

MAX_LEN = 50

def encode_sentence(sent):
    tokens = [word_vocab.get(t, 1) for t, _ in sent][:MAX_LEN]
    labels = [LABEL2ID[l]          for _, l in sent][:MAX_LEN]
    pad    = MAX_LEN - len(tokens)
    return tokens + [0]*pad, labels + [0]*pad


class NERDataset(Dataset):
    def __init__(self, sentences):
        self.data = [encode_sentence(s) for s in sentences]
    def __len__(self): return len(self.data)
    def __getitem__(self, idx):
        t, l = self.data[idx]
        return torch.tensor(t), torch.tensor(l)


class BiLSTMNER(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim, output_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        self.lstm      = nn.LSTM(emb_dim, hidden_dim, bidirectional=True,
                                 batch_first=True)
        self.fc        = nn.Linear(hidden_dim * 2, output_dim)
    def forward(self, x):
        return self.fc(self.lstm(self.embedding(x))[0])


def train_bilstm(train_sents, test_sents, epochs=3):
    train_loader = DataLoader(NERDataset(train_sents), batch_size=32, shuffle=True)
    test_loader  = DataLoader(NERDataset(test_sents),  batch_size=32)

    model     = BiLSTMNER(len(word_vocab), 128, 128, len(LABELS)).to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    model.train()
    for epoch in range(epochs):
        total = 0
        for tokens, labels in tqdm(train_loader, desc=f"    Epoch {epoch+1}",
                                   leave=False):
            tokens, labels = tokens.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            loss = criterion(model(tokens).view(-1, len(LABELS)), labels.view(-1))
            loss.backward()
            optimizer.step()
            total += loss.item()
        print(f"    Epoch {epoch+1} Loss: {total:.4f}")

    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for tokens, labels in test_loader:
            preds  = torch.argmax(model(tokens.to(DEVICE)), dim=-1).cpu().numpy()
            labels = labels.numpy()
            for p_seq, l_seq in zip(preds, labels):
                y_pred.append([LABELS[p] for p in p_seq])
                y_true.append([LABELS[l] for l in l_seq])
    return evaluate(y_true, y_pred)


# ============================================================
# TRANSFORMER BASELINE  (mBERT fine-tuned for NER)
# ============================================================
#
# Model: bert-base-multilingual-cased
#   - Covers 104 languages including Bengali
#   - Pre-trained on Wikipedia; fine-tuned here on WikiANN Bengali
#   - Token classification head: Linear(768, n_labels) over each subword
#
# Subword alignment:
#   BERT tokenizes into subwords (wordpieces). Each original token may
#   map to multiple subwords. We assign the NER label only to the FIRST
#   subword of each token and mask the rest with -100 so they are
#   ignored by CrossEntropyLoss. At evaluation time we recover
#   one prediction per original token using the offset mapping.
#
# Training details:
#   - AdamW optimizer with linear warmup schedule
#   - Max sequence length: 128 subword tokens
#   - Batch size: 16 (smaller than BiLSTM due to GPU memory)
#   - 3 epochs (same as BiLSTM for fair comparison)
# ============================================================

TRANSFORMER_MODEL   = "bert-base-multilingual-cased"
TRANSFORMER_MAX_LEN = 128


class TransformerNERDataset(Dataset):
    """
    Tokenizes sentences with a HuggingFace tokenizer and aligns NER labels
    to subword tokens. Labels for non-first subwords are set to -100
    (ignored by loss). Returns input_ids, attention_mask, labels.
    """

    def __init__(self, sentences, tokenizer):
        self.tokenizer = tokenizer
        self.sentences = sentences

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, idx):
        sent   = self.sentences[idx]
        words  = [tok for tok, _   in sent]
        labels = [lbl for _,   lbl in sent]

        encoding = self.tokenizer(
            words,
            is_split_into_words=True,
            truncation=True,
            max_length=TRANSFORMER_MAX_LEN,
            padding="max_length",
            return_tensors="pt",
        )

        word_ids       = encoding.word_ids()
        aligned_labels = []
        prev_word_id   = None
        for wid in word_ids:
            if wid is None:
                aligned_labels.append(-100)
            elif wid != prev_word_id:
                aligned_labels.append(LABEL2ID[labels[wid]])
            else:
                aligned_labels.append(-100)
            prev_word_id = wid

        return {
            "input_ids":      encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels":         torch.tensor(aligned_labels, dtype=torch.long),
        }


def train_transformer(train_sents, test_sents, epochs=3):
    """
    Fine-tune mBERT for Bengali NER token classification.
    Returns evaluate() dict with Precision, Recall, F1, Accuracy.
    """
    from transformers import AutoTokenizer, AutoModelForTokenClassification
    from transformers import get_linear_schedule_with_warmup

    print(f"    Loading tokenizer: {TRANSFORMER_MODEL}")
    tokenizer = AutoTokenizer.from_pretrained(TRANSFORMER_MODEL)

    train_ds = TransformerNERDataset(train_sents, tokenizer)
    test_ds  = TransformerNERDataset(test_sents,  tokenizer)

    train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
    test_loader  = DataLoader(test_ds,  batch_size=16)

    model = AutoModelForTokenClassification.from_pretrained(
        TRANSFORMER_MODEL,
        num_labels=len(LABELS),
        ignore_mismatched_sizes=True,
    ).to(DEVICE)

    optimizer   = torch.optim.AdamW(model.parameters(), lr=2e-5, weight_decay=0.01)
    total_steps = len(train_loader) * epochs
    scheduler   = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=int(0.1 * total_steps),
        num_training_steps=total_steps,
    )

    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for batch in tqdm(train_loader, desc=f"    Epoch {epoch+1}", leave=False):
            input_ids      = batch["input_ids"].to(DEVICE)
            attention_mask = batch["attention_mask"].to(DEVICE)
            labels         = batch["labels"].to(DEVICE)

            optimizer.zero_grad()
            outputs = model(input_ids=input_ids,
                            attention_mask=attention_mask,
                            labels=labels)
            outputs.loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            total_loss += outputs.loss.item()
        print(f"    Epoch {epoch+1} Loss: {total_loss:.4f}")

    # Evaluation: recover one prediction per original token (first subword only)
    model.eval()
    y_true_all, y_pred_all = [], []

    with torch.no_grad():
        for batch in test_loader:
            input_ids      = batch["input_ids"].to(DEVICE)
            attention_mask = batch["attention_mask"].to(DEVICE)
            label_ids      = batch["labels"]

            logits = model(input_ids=input_ids,
                           attention_mask=attention_mask).logits
            preds  = torch.argmax(logits, dim=-1).cpu()

            for i in range(input_ids.size(0)):
                true_seq, pred_seq = [], []
                for t in range(label_ids.size(1)):
                    true_id = label_ids[i, t].item()
                    if true_id == -100:
                        continue
                    true_seq.append(LABELS[true_id])
                    pred_seq.append(LABELS[preds[i, t].item()])
                if true_seq:
                    y_true_all.append(true_seq)
                    y_pred_all.append(pred_seq)

    return evaluate(y_true_all, y_pred_all)



# ============================================================
# EVALUATION & UTILITIES
# ============================================================

def evaluate(y_true, y_pred):
    return {
        "Precision": precision_score(y_true, y_pred),
        "Recall":    recall_score(y_true, y_pred),
        "F1":        f1_score(y_true, y_pred),
        "Accuracy":  accuracy_score(y_true, y_pred),
    }

def evaluate_crf(crf, X_test, y_test):
    return evaluate(y_test, crf.predict(X_test))

def create_subset(data, pct):
    return random.sample(data, int(len(data) * pct))


# ============================================================
# ABLATION STUDY
# ============================================================
#
# Removes one feature group at a time from the full Latent CRF.
# Run at 10% training data.
#
# Feature groups:
#   morph   — MAP stem/suffix, top-1/2 suffix, top1 prob, entropy
#   ngrams  — character prefix/suffix n-grams (length 2 and 3)
#   context — previous and next word features (word, suffix, n-gram)
#   word    — word identity and lowercase form
#
# The F1 drop when a group is removed = that group's contribution.
# ============================================================

ABLATION_CONFIGS = {
    "Full Model":       set(),
    "− Morphology":     {"morph"},
    "− Char N-grams":   {"ngrams"},
    "− Context":        {"context"},
    "− Word Identity":  {"word"},
}


def run_ablation(train_subset, X_test_latent_fn, y_test, n_em_iters=3):
    """
    Run ablation study: train Latent CRF with each feature group removed.
    Returns list of result dicts.
    """
    print("\n[ABLATION STUDY — 10% training data]")
    results = []

    for config_name, ablate_set in ABLATION_CONFIGS.items():
        print(f"\n  Config: {config_name}")
        crf, segmenter = em_train_latent_crf(train_subset,
                                              n_em_iters=n_em_iters,
                                              ablate=ablate_set)
        X_test = X_test_latent_fn(segmenter, ablate_set)
        res    = evaluate_crf(crf, X_test, y_test)

        drop = None  # filled in after Full Model is known
        results.append({
            "Config":    config_name,
            "F1":        res["F1"],
            "Precision": res["Precision"],
            "Recall":    res["Recall"],
            "ablate":    ablate_set,
        })
        print(f"    F1 = {res['F1']:.4f}")

    # Compute F1 drop relative to Full Model
    full_f1 = next(r["F1"] for r in results if r["Config"] == "Full Model")
    for r in results:
        r["F1 Drop"] = full_f1 - r["F1"]

    return results


# ============================================================
# MAIN EXPERIMENT
# ============================================================

PERCENTAGES = [0.01, 0.05, 0.10, 0.30, 0.50, 1.00]
EM_ITERS    = 3

main_results    = []
ablation_subset = None   # 10% subset, reused for ablation
final_segmenter = None   # from 100% run
ABLATION_PCT    = 0.10   # ablation study uses 10% training data

print("\nPreparing basic test features...")
X_test_basic, y_test = prepare_data_basic(test_sentences)

for p in PERCENTAGES:
    pct_label = int(p * 100)
    print(f"\n{'='*60}")
    print(f"  TRAINING WITH {pct_label}% DATA")
    print(f"{'='*60}")

    subset = create_subset(train_sentences, p)
    if p == ABLATION_PCT:
        ablation_subset = subset   # save 10% subset for ablation

    # ── 1. Basic CRF ──────────────────────────────────────────
    print("\n[1/4] Basic CRF")
    X_tr_b, y_tr_b = prepare_data_basic(subset)
    basic_crf = train_crf(X_tr_b, y_tr_b)
    basic_res = evaluate_crf(basic_crf, X_test_basic, y_test)
    print(f"  F1: {basic_res['F1']:.4f}  Accuracy: {basic_res['Accuracy']:.4f}")

    # ── 2. Latent Morphology CRF ──────────────────────────────
    print(f"\n[2/4] Latent Morphology CRF  (EM x {EM_ITERS})")
    latent_crf, segmenter = em_train_latent_crf(subset, n_em_iters=EM_ITERS)

    X_test_latent = [
        [word2features_latent(sent, i, segmenter) for i in range(len(sent))]
        for sent in test_sentences
    ]
    latent_res = evaluate_crf(latent_crf, X_test_latent, y_test)
    print(f"  F1: {latent_res['F1']:.4f}  Accuracy: {latent_res['Accuracy']:.4f}")

    # ── 3. BiLSTM ─────────────────────────────────────────────
    print("\n[3/4] BiLSTM")
    bilstm_res = train_bilstm(subset, test_sentences)
    print(f"  F1: {bilstm_res['F1']:.4f}  Accuracy: {bilstm_res['Accuracy']:.4f}")

    # ── 4. Transformer (mBERT) ────────────────────────────────
    print("\n[4/4] Transformer (mBERT)")
    transformer_res = train_transformer(subset, test_sentences)
    print(f"  F1: {transformer_res['F1']:.4f}  Accuracy: {transformer_res['Accuracy']:.4f}")

    main_results.append({
        "Dataset %":            pct_label,
        "Basic CRF F1":         basic_res["F1"],
        "Basic CRF Prec":       basic_res["Precision"],
        "Basic CRF Rec":        basic_res["Recall"],
        "Basic CRF Acc":        basic_res["Accuracy"],
        "Latent CRF F1":        latent_res["F1"],
        "Latent CRF Prec":      latent_res["Precision"],
        "Latent CRF Rec":       latent_res["Recall"],
        "Latent CRF Acc":       latent_res["Accuracy"],
        "BiLSTM F1":            bilstm_res["F1"],
        "BiLSTM Prec":          bilstm_res["Precision"],
        "BiLSTM Rec":           bilstm_res["Recall"],
        "BiLSTM Acc":           bilstm_res["Accuracy"],
        "mBERT F1":             transformer_res["F1"],
        "mBERT Prec":           transformer_res["Precision"],
        "mBERT Rec":            transformer_res["Recall"],
        "mBERT Acc":            transformer_res["Accuracy"],
    })

    if p == 1.00:
        final_segmenter = segmenter


# ============================================================
# ABLATION STUDY (10% data)
# ============================================================

def make_latent_test_features(segmenter, ablate=None):
    if ablate is None:
        ablate = set()
    return [
        [word2features_latent(sent, i, segmenter, ablate)
         for i in range(len(sent))]
        for sent in test_sentences
    ]

ablation_results = run_ablation(
    ablation_subset,
    make_latent_test_features,
    y_test,
    n_em_iters=EM_ITERS
)


# ============================================================
# RESULTS TABLE & CSV
# ============================================================

results_df   = pd.DataFrame(main_results)
ablation_df  = pd.DataFrame(ablation_results).drop(columns=["ablate"])

print("\n\nFINAL RESULTS")
print(results_df[[
    "Dataset %",
    "Basic CRF F1",  "Basic CRF Acc",
    "Latent CRF F1", "Latent CRF Acc",
    "BiLSTM F1",     "BiLSTM Acc",
    "mBERT F1",      "mBERT Acc",
]].to_string(index=False))

print("\n\nABLATION STUDY (10% data)")
print(ablation_df[["Config", "F1", "Precision", "Recall", "F1 Drop"]]
      .to_string(index=False))

results_df.to_csv("final_results_latent.csv", index=False)
ablation_df.to_csv("ablation_results.csv",    index=False)
print("\nSaved to final_results_latent.csv and ablation_results.csv")


# ============================================================
# PLOTS
# ============================================================

fig = plt.figure(figsize=(18, 6))
gs  = gridspec.GridSpec(1, 2, figure=fig, wspace=0.35)

# ── Plot 1: F1 vs Training Data % ───────────────────────────
ax1 = fig.add_subplot(gs[0, 0])
for col, label, marker, color in [
    ("Basic CRF F1",  "Basic CRF",         "o", "#2196F3"),
    ("Latent CRF F1", "Latent Morph CRF",  "^", "#4CAF50"),
    ("BiLSTM F1",     "BiLSTM",            "D", "#F44336"),
    ("mBERT F1",      "mBERT",             "s", "#9C27B0"),
]:
    ax1.plot(results_df["Dataset %"], results_df[col],
             marker=marker, color=color, linewidth=2,
             markersize=7, label=label)

ax1.set_xlabel("Training Data (%)", fontsize=12)
ax1.set_ylabel("F1 Score",          fontsize=12)
ax1.set_title("Model Comparison\nBasic CRF vs Latent Morph CRF vs BiLSTM vs mBERT",
              fontsize=11)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.35)
ax1.set_ylim(0, 1.0)

# ── Plot 1b (inset/twin axis): Accuracy vs Training Data % ──
ax1b = ax1.twinx()
for col, label, color, ls in [
    ("Basic CRF Acc",  "Basic CRF Acc",  "#2196F3", ":"),
    ("Latent CRF Acc", "Latent CRF Acc", "#4CAF50", ":"),
    ("BiLSTM Acc",     "BiLSTM Acc",     "#F44336", ":"),
    ("mBERT Acc",      "mBERT Acc",      "#9C27B0", ":"),
]:
    ax1b.plot(results_df["Dataset %"], results_df[col],
              color=color, linewidth=1.2, linestyle=ls, alpha=0.6)
ax1b.set_ylabel("Accuracy (dotted)", fontsize=10, color="grey")
ax1b.tick_params(axis='y', labelcolor='grey')
ax1b.set_ylim(0, 1.0)

# ── Plot 2: Ablation Study bar chart ────────────────────────
ax2 = fig.add_subplot(gs[0, 1])

configs = [r["Config"] for r in ablation_results]
f1s     = [r["F1"]     for r in ablation_results]
drops   = [r["F1 Drop"] for r in ablation_results]

# Color: Full Model in green, ablations in shades of orange/red by drop size
colors = []
for r in ablation_results:
    if r["Config"] == "Full Model":
        colors.append("#4CAF50")
    elif r["F1 Drop"] > 0.02:
        colors.append("#F44336")   # large drop — red
    elif r["F1 Drop"] > 0.005:
        colors.append("#FF9800")   # medium drop — orange
    else:
        colors.append("#FFC107")   # small drop — yellow

bars = ax2.barh(configs, f1s, color=colors, alpha=0.88, edgecolor="white")

# Annotate bars with F1 and drop
full_f1 = next(r["F1"] for r in ablation_results if r["Config"] == "Full Model")
for bar, r in zip(bars, ablation_results):
    drop_str = (f"  F1={r['F1']:.4f}"
                if r["Config"] == "Full Model"
                else f"  F1={r['F1']:.4f}  (−{r['F1 Drop']:.4f})")
    ax2.text(bar.get_width() + 0.002, bar.get_y() + bar.get_height() / 2,
             drop_str, va='center', fontsize=9)

ax2.set_xlabel("F1 Score", fontsize=12)
ax2.set_title("Feature Ablation Study\n(Latent Morph CRF @ 10% data)",
              fontsize=11)
ax2.set_xlim(0, 1.05)
ax2.axvline(full_f1, color="#4CAF50", linestyle="--", linewidth=1.2,
            alpha=0.7, label="Full Model F1")
ax2.legend(fontsize=9)
ax2.grid(True, axis='x', alpha=0.35)
ax2.invert_yaxis()   # Full Model at top

fig.suptitle("Bengali NER: Latent Morphology CRF Analysis",
             fontsize=13, fontweight='bold')
plt.savefig("bengali_ner_results.png", dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved to bengali_ner_results.png")


# ============================================================
# PRINTED SUMMARIES
# ============================================================

# ── Best model per data size ──────────────────────────────────
print("\n\nBEST MODEL PER DATA SIZE\n")
model_cols = {
    "Basic CRF":        ("Basic CRF F1",  "Basic CRF Acc"),
    "Latent Morph CRF": ("Latent CRF F1", "Latent CRF Acc"),
    "BiLSTM":           ("BiLSTM F1",     "BiLSTM Acc"),
    "mBERT":            ("mBERT F1",      "mBERT Acc"),
}
for _, row in results_df.iterrows():
    scores = {m: row[cols[0]] for m, cols in model_cols.items()}
    best   = max(scores, key=scores.get)
    accs   = {m: row[cols[1]] for m, cols in model_cols.items()}
    print(f"  {int(row['Dataset %']):3d}% => {best:22s}  "
          f"(F1={scores[best]:.4f}, Acc={accs[best]:.4f})")

print("\n\nFULL ACCURACY TABLE\n")
print(f"  {'Dataset %':>10}  {'Basic CRF':>10}  {'Latent CRF':>10}  "
      f"{'BiLSTM':>10}  {'mBERT':>10}")
for _, row in results_df.iterrows():
    print(f"  {int(row['Dataset %']):>10d}  "
          f"{row['Basic CRF Acc']:>10.4f}  "
          f"{row['Latent CRF Acc']:>10.4f}  "
          f"{row['BiLSTM Acc']:>10.4f}  "
          f"{row['mBERT Acc']:>10.4f}")

# ── Ablation ranked by contribution ──────────────────────────
print("\n\nFEATURE CONTRIBUTION RANKING (10% data)\n")
ranked = sorted(
    [r for r in ablation_results if r["Config"] != "Full Model"],
    key=lambda x: -x["F1 Drop"]
)
print(f"  Full Model F1 = {full_f1:.4f}\n")
print(f"  {'Feature Group':<20}  {'F1 without':>10}  {'F1 Drop':>9}")
for r in ranked:
    group = r["Config"].replace("− ", "")
    print(f"  {group:<20}  {r['F1']:>10.4f}  {r['F1 Drop']:>+9.4f}")

# ── Learned morphological patterns ───────────────────────────
if final_segmenter is not None:
    print("\n\nLEARNED MORPHOLOGICAL PATTERNS  (100% data)\n")
    seen = {k: v for k, v in final_segmenter.log_suffix_weights.items()
            if k in final_segmenter._seen_suffixes}
    top10 = sorted(seen.items(), key=lambda x: -x[1])[:10]
    print("Top 10 suffixes (within-bucket probability):")
    for sfx, logw in top10:
        label = sfx if sfx else "∅ (empty)"
        print(f"  '{label}' => p={np.exp(logw):.4f}")

    print("\nExample segmentations on test tokens:")
    example = [tok for sent in test_sentences[:5] for tok, _ in sent][:15]
    for token in example:
        stem, suffix = final_segmenter.most_likely_segmentation(token)
        post         = final_segmenter.posterior(token)
        entropy      = -sum(p * np.log(p + 1e-12) for p in post.values())
        print(f"  {token:20s} => stem='{stem}' | suffix='{suffix}' "
              f"| H={entropy:.2f}")
```
